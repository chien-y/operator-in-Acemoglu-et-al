```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```


The clearing repayment satisfies 
\begin{equation}
\mathbf p = [(\mathbf{p \Pi + e } + \zeta A \mathbf{1}) \wedge \bar{\mathbf p}]^+
\end{equation}
where $\mathbf e = \mathbf c + \mathbf z - \nu$.
$\mathbf c$ denotes cash holding, which is normalized to be $\mathbf 0$.
$\mathbf z$ is the project return, which is volatile.
$\nu$ is the sinior liability.

Assume that $z_j \in \{a, a-\epsilon \}$ with $\epsilon \in (a - \nu + \zeta A , a)$ and $a > \nu$.

Thus, the cash flow is negative $e_j + \zeta A \in (-\nu + \zeta A, 0)$ if $z_j = a - \epsilon$.

The cash flow is positive $e_j + \zeta A = a + \zeta A$ if the bank is not distressed by negative shocks.

## The operator on component $i$, fixing the other components.

Define $\Phi: [0, \mathbf{\bar p}] \rightarrow [0, \mathbf{\bar p}]$ as 
$\Phi \mathbf p = [(\mathbf{p \Pi + e } + \zeta A \mathbf{1}) \wedge \bar{\mathbf p}^+$.

$ \Phi \mathbf p(i)$ is not depends on $p(i)$ when $\pi^{ii}=0$, following Eisenberg and Noe.

Let $d_i =  \sum_{j \neq i} p(j)\Pi(j, i) + e_i + \zeta A$. 
Then $ \Phi \mathbf p(i) = (d_i)^+ \wedge \mathbf{\bar p}(i)$.

Assume bank $i$ is distressed by negative shock.
So, $- \nu + \zeta A < d_i < (\mathbf{\bar p} \Pi) (i)$. 

Fix $p(j)$ for every $j \neq i$ so that $d_i$ is fixed.
Then $d_i > 0$ if the bank is not distressed, while $d_i$ may be negative if the bank has a negative shock.

The operator for bank $i$ is $p(i) = (d_i)^+ \wedge \bar p(i)$

Clearly, the fixed point is unique given $d_i$. It is $1$ if $d_i > \bar p(i)$ and $0$ if $d_i < 0$


In the plot, suppose that $\zeta = 0$ and let $g(i) = (\mathbf{\bar p} \Pi) (i)$


```python
p_bar = 1
ν = 0.5
g_i = 1
d_i = np.random.random() * (g_i + ν) - ν
print("d_i is ", d_i)
Φp = np.minimum(np.maximum(d_i, 0), p_bar)

x = np.linspace(-0.5, 1.2)

plt.figure(figsize=(5, 5))
plt.plot(x, x, 'k--', alpha=0.5)
plt.hlines(Φp, 0, 1, alpha=0.8, label='Φp')

if Φp >= p_bar:
    plt.markers(1, 1,'r.', markersize=10, label='fixed point')
elif Φp <= 0:
    plt.plot(0, 0,'r.', markersize=10, label='fixed point')
else:
    plt.plot(Φp, Φp, c='r', marker='.', markersize=10, label='fixed point')
    
plt.legend()
plt.show()
```

## The operator $\Phi p = (p + b)^+ \wedge \bar p$.

Suppose that there is only one node, index $1$.

Let $\pi^{11}=1$ for a stochastic matrix $\Pi$.
Then the operator in one dimension is $\Phi p = (p + b)^+ \wedge \bar p$, where $p, b \bar p$ are scalars.

Consider a negative shock, $b < 0$.

```python
p_bar = 1
ν = 0.5
p = np.linspace(0,1, endpoint=True)
b = np.random.random() * ν - ν
Φp = np.minimum(np.maximum(p + b, 0), p_bar)

def plot_Φp(p, Φp, b):    
    plt.plot(p, Φp, label='Φp')
    plt.plot(p, p, 'k--', alpha=0.4)
    if b <= 0:
        plt.plot(0, 0, 'r.', markersize=10, label='fixed point')
    else:
        plt.plot(1, 1, 'r.', markersize=10, label='fixed point')
    plt.legend()    
    plt.show()
plot_Φp(p, Φp, b)
print('total cash flow is ', b)
```

Consider $b > 0$. That is, $b = a - \nu + \zeta A$.

```python
b = 0.1
Φp = np.minimum(np.maximum(p + b, 0), p_bar)
plot_Φp(p, Φp, b)
```
